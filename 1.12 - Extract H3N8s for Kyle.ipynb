```python
data = pd.read_csv('Raw Data/20140211 Minto and North America Combined Dates Cleaned.csv', index_col=0)
data = data[data['Subtype'] == 'H3N8']
data.rename(columns={'Sequence Accession':'Sequence Accession1'}, inplace=True)
data['Collection Date'] = data['Collection Date'].str.split(' ').str[0]
data['Taxa Name'] = data['Strain Name'] + '|' + data['Subtype'] + '|' + data['Collection Date']
data['Taxa Name'] = data['Taxa Name'].str.replace(' ', '_')
data['Taxa Name'] = data['Taxa Name'].str.replace('-', '/')
len(data)
```

```python
# Create a dictionary of accession numbers and taxa names
accession_taxa = dict()
for i in range(1,9):
    accession_taxa[i] = dict(zip(data['Sequence Accession{0}'.format(i)], data['Taxa Name']))
```

```python
# Read the FASTA files into memory
segments_dict = dict()
for i in range(1, 9):
    segments_dict[i] = SeqIO.to_dict(SeqIO.parse('1.0 - Split FASTA Files/Segment {0}.fasta'.format(i), 'fasta'))
```

```python
# Delete anything that isn't H3N8.
for i in range(1, 9):
    for accession in segments_dict[i].keys():
        if accession not in data['Sequence Accession{0}'.format(i)].values:
            del segments_dict[i][accession]
```

```python
# Function to find longest ORFs:
def longest_orfs(list_of_seqrecords):
    aa_sequences = []
    for i, record in enumerate(list_of_seqrecords):
        longest_protein = SeqRecord(id=record.id, seq='')
        for frame in range(3):
            length = 3 * ((len(record) - frame) // 3)
            translation = record.seq[frame:frame + length].translate()
            for pro in translation.split("*"):
                if len(pro) > len(longest_protein.seq):
                    longest_protein.seq = pro
                    
        longest_protein.seq = longest_protein.seq[str(longest_protein.seq).index('M'):]
        aa_sequences.append(longest_protein)
        
    return aa_sequences
```

```python
# Identify the longest ORF
orfs = dict()
for i in range(1, 9):
    orfs[i] = longest_orfs(segments_dict[i].values())
```

```python
# Compute the ORF lengths
orf_lengths = dict()
for i in range(1, 9):
    orf_lengths[i] = list()
    for seqrecord in orfs[i]:
        orf_lengths[i].append(len(seqrecord.seq))
```

```python
# Plot the histogram of ORF lengths
for i in range(1, 9):
    fig = plt.figure(i)
    plt.hist(orf_lengths[i])
```

```python
# Convert the names of each sequence ID into the taxa name.
for i in range(1,9):
    for accession, seqrecord in segments_dict[i].items():
        seqrecord.id = accession_taxa[i][accession]
        seqrecord.description = ''
```

```python
# Write the sequence files to disk
for i in range(1,9):
    SeqIO.write(segments_dict[i].values(), '1.12 - H3N8 Sequences/H3N8 Segment {0}.fasta'.format(i), 'fasta')
```
