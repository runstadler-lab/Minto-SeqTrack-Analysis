```python
# Clean out sequences where there may be potential problems:
# - has asterisks in sequence
# - subtypes are mixed
# - subtypes are unclear

dropped = {}

for row in joined.iterrows():
    print(row[0])
    if row[1]['Subtype'] == 'Mixed':
        dropped[row[1]['Strain Name']] = "mixed"
        joined = joined.drop(row[0])
        
    elif row[1]['Subtype'] == 'mixed':
        dropped[row[1]['Strain Name']] = "mixed"
        joined = joined.drop(row[0])
            
    elif '*' in row[1]['Sequence Accession']:
        dropped[row[1]['Strain Name']] = "bad1"
        joined = joined.drop(row[0])
        
    elif '*' in row[1]['Sequence Accession2']:
        dropped[row[1]['Strain Name']] = "bad2"
        joined = joined.drop(row[0])
        
    elif '*' in row[1]['Sequence Accession3']:
        dropped[row[1]['Strain Name']] = "bad3"
        joined = joined.drop(row[0])
        
    elif '*' in row[1]['Sequence Accession4']:
        dropped[row[1]['Strain Name']] = "bad4"
        joined = joined.drop(row[0])
    
    elif '*' in row[1]['Sequence Accession5']:
        dropped[row[1]['Strain Name']] = "bad5"
        joined = joined.drop(row[0])

    elif '*' in row[1]['Sequence Accession6']:
        dropped[row[1]['Strain Name']] = "bad6"
        joined = joined.drop(row[0])
        
    elif '*' in row[1]['Sequence Accession7']:
        dropped[row[1]['Strain Name']] = "bad7"
        joined = joined.drop(row[0])
        
    elif '*' in row[1]['Sequence Accession8']:
        dropped[row[1]['Strain Name']] = "bad8"
        joined = joined.drop(row[0])
```

```python
pd.DataFrame(zip(dropped.keys(), dropped.values())).to_csv('Raw Data/20140924 Dropped Isolates and Reasons.csv')
```
