```python
import pandas as pd
```

```python
minto = pd.read_csv('Raw Data/20140206 Minto Flats 520 with Accessions.csv')
minto = minto[['Common Name', 'Sequence Accession', 'Sequence Accession2', 'Sequence Accession3',
               'Sequence Accession4', 'Sequence Accession5', 'Sequence Accession6', 'Sequence Accession7',
               'Sequence Accession8', 'Date', 'Species', 'Subtype']]
#minto = minto.set_index('Sequence Accession')
minto['Strain Name'] = minto['Common Name']
del minto['Common Name']

minto['Host Species'] = minto['Species']
del minto['Species']


minto['Collection Date'] = minto['Date']
del minto['Date']
minto
```

```python
america = pd.read_csv('Raw Data/20140206 Minto Plus North America with Accession.csv')
#america = america.set_index('Sequence Accession')
len(america)
```

```python
joined = america.append(minto)
print len(joined)
joined
```

```python
# Clean out sequences where there may be potential problems:
# - has asterisks in sequence
# - subtypes are mixed
# - subtypes are unclear

for row in joined.iterrows():
    if row[1]['Subtype'] == 'Mixed':
        joined = joined.drop(row[0])
        
    elif row[1]['Subtype'] == 'mixed':
        joined = joined.drop(row[0])
        
#     elif r"(" in row[1]['Subtype']:
#         joined = joined.drop(row[0])
    
    elif '*' in row[1]['Sequence Accession']:
        joined = joined.drop(row[0])
        
    elif '*' in row[1]['Sequence Accession2']:
        joined = joined.drop(row[0])
        
    elif '*' in row[1]['Sequence Accession3']:
        joined = joined.drop(row[0])
        
    elif '*' in row[1]['Sequence Accession4']:
        joined = joined.drop(row[0])
    
    elif '*' in row[1]['Sequence Accession5']:
        joined = joined.drop(row[0])

    elif '*' in row[1]['Sequence Accession6']:
        joined = joined.drop(row[0])
        
    elif '*' in row[1]['Sequence Accession7']:
        joined = joined.drop(row[0])
        
    elif '*' in row[1]['Sequence Accession8']:
        joined = joined.drop(row[0])
```

```python
joined = joined.drop_duplicates(subset=['Sequence Accession', 'Sequence Accession2', 'Sequence Accession3', 'Sequence Accession4', 
                                      'Sequence Accession5', 'Sequence Accession6', 'Sequence Accession7', 'Sequence Accession8'])
print len(joined)
joined.to_csv('Raw Data/20140211 Minto and North America Combined.csv')
```
