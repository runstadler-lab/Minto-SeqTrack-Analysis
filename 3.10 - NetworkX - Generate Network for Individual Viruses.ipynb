```python
import networkx as nx
import numpy as np
import pandas as pd
from networkx.readwrite import json_graph
import json
import datetime
```

```python
data = pd.read_csv('3.0 - Transmissions/Minto Plus North America Full Transmissions.csv', index_col=0, parse_dates=['Collection Date1', 'Collection Date2'])
data['Collection Date1'] = pd.to_datetime(data['Collection Date1'])
data['Collection Date2'] = pd.to_datetime(data['Collection Date2'])
data.dtypes
```

```python
data
```

```python
G = nx.DiGraph()
```

```python
from dateutil.tz import tzutc

UTC = tzutc()

def serialize_date(dt):
    """
    Serialize a date/time value into an ISO8601 text representation
    adjusted (if needed) to UTC timezone.

    For instance:
    >>> serialize_date(datetime(2012, 4, 10, 22, 38, 20, 604391))
    '2012-04-10T22:38:20.604391Z'
    """
    #if dt.tzinfo:
    #    dt = dt.astimezone(UTC)
    return dt.isoformat()
```

```python
for row in data.iterrows():
#     print row[1]
    dsplit1 = str(row[1]['Collection Date1']).split(' ')[0].split('-')
    dsplit2 = str(row[1]['Collection Date2']).split(' ')[0].split('-')
    
    G.add_node(row[1]['Isolate1'], 
               host=row[1]['Host Species1'], 
               isolation_date=serialize_date(datetime.date(int(dsplit1[0]), int(dsplit1[1]), int(dsplit1[2]))),
               subtype=row[1]['Subtype1'],
               country=row[1]['Country1'],
               state=row[1]['State/Province1'])
    G.add_node(row[1]['Isolate2'], 
               host=row[1]['Host Species2'], 
               isolation_date=serialize_date(datetime.date(int(dsplit2[0]), int(dsplit2[1]), int(dsplit2[2]))),
               subtype=row[1]['Subtype2'],
               country=row[1]['Country2'],
               state=row[1]['State/Province2'])
    
    if row[1]['Host Species1'] == row[1]['Host Species2']:
        G.add_edge(row[1]['Isolate1'], 
                   row[1]['Isolate2'], 
                   time_delta=row[1]['Time Delta'], 
                   weight=row[1]['PWI'], 
                   trtype='Intra-Species')
    if row[1]['Host Species1'] != row[1]['Host Species2']:
        G.add_edge(row[1]['Isolate1'], 
                   row[1]['Isolate2'], 
                   time_delta=row[1]['Time Delta'], 
                   weight=row[1]['PWI'], 
                   trtype='Inter-Species')
```

```python
for node in G.nodes():
    edges = G.in_edges(node, data=True)
    if len(edges) > 0:
        max_weight = max([edge[2]['weight'] for edge in edges])
        for edge in edges:
            if edge[2]['weight'] < max_weight:
                G.remove_edge(edge[0], edge[1])
```

```python
from datetime import timedelta
for edge in G.edges(data=True):
    if len(edge[2]['time_delta'].split(',')) == 2:
        days = edge[2]['time_delta'].split(',')[0].split(' ')[0]
        days = int(days)
    else:
        days = 0
    
    if timedelta(days) > timedelta(15):
        edge[2]['tr_timeclass'] = "Environmental Persistence"
    if timedelta(days) <= timedelta(15):
        edge[2]['tr_timeclass'] = "Direct Transmission"
```

```python
d = json_graph.node_link_data(G) # node-link format to serialize
json.dump(d, open('viz/force2.json','w'))
```

```python
# Write CSV file of the graph

df_edges = pd.DataFrame()

for edge in G.edges(data=True):
    df_edge = pd.Series()
    df_edge['Source'] = edge[0]
    df_edge['Sink'] = edge[1]
    df_edge['Transmission Type'] = edge[2]['trtype']

    if len(str(edge[2]['time_delta']).split(',')) == 2:
        df_edge['Time Delta'] = str(edge[2]['time_delta']).split(',')[0].split(' ')[0]

    if len(str(edge[2]['time_delta']).split(',')) == 1:
        df_edge['Time Delta'] = '0'
        
#     df_edge['Time Delta'] = edge[2]['time_delta']
    df_edge['Weight'] = edge[2]['weight']
    df_edge['Time Class'] = edge[2]['tr_timeclass']
   
    df_edges = df_edges.append(df_edge, ignore_index=True)
    
df_edges
```

```python
df_edges.to_csv('3.10 - Full Transmission as CSV/Full Transmissions Edges.csv')
```

```python
df_nodes = pd.DataFrame()

for node in G.nodes(data=True):
    node_series = pd.Series()
#     print node
    node_series['Isolate Name'] = node[0]
    node_series['Country'] = node[1]['country']
    node_series['State'] = node[1]['state']
    node_series['Subtype'] = node[1]['subtype']
    node_series['Host'] = node[1]['host']
    node_series['Isolation Date'] = node[1]['isolation_date']
    
    df_nodes = df_nodes.append(node_series, ignore_index=True)
    
df_nodes
```

```python
df_nodes.to_csv('3.10 - Full Transmission as CSV/Full Transmissions Nodes.csv')
```
