```python
# Load the data
data = pd.read_csv('3.0 - Transmissions/Minto Plus North America All Transmissions.csv', index_col=0, parse_dates=['Collection Date1', 'Collection Date2'])
data['Collection Date1'] = pd.to_datetime(data['Collection Date1'])
data['Collection Date2'] = pd.to_datetime(data['Collection Date2'])
data['Time Delta'] = data['Collection Date2'] - data['Collection Date1']
data['Time Delta'] = data['Time Delta'].astype('timedelta64[D]')
data
```

```python
# Initialize the graph that will be used to store the transmission network
G = nx.DiGraph()

# Initialize a graph that will hold all of the discarded edges from G. This is used to
# add back edges that are sources for isolates that may have been implicated in bi-directional
# reconstructions as a result of the "-6 to +15" rule.
G_discarded = nx.DiGraph()
```

```python
# Add nodes into the graph, along with metadata.
for group in data.groupby(['Isolate1', 'Collection Date1', 'Subtype1', 'Host Species1','State/Province1']):
#     dsplit = str(group[0][1]).split(' ')[0].split('-')
#     isolate_date = serialize_date(datetime(int(dsplit[0]), int(dsplit[1]), int(dsplit[2])))
    G.add_node(group[0][0],
               isolation_date=group[0][1],
               subtype=group[0][2],
               host=group[0][3],
               state=group[0][4])
```

```python
# Add in edges to the graph, and classify them as intra-species or inter-species.
for row in data[data['PWI'] > 7.0].iterrows():
    # 12 September 2014: I just noticed here that 
    #     if Time Delta >= -7 
    # should have been:
    #     if Time Delta >= -6.
    # Made the change here.
    if row[1]['Time Delta'] >= -6:
        if row[1]['Host Species1'] == row[1]['Host Species2']:
            G.add_edge(row[1]['Isolate1'], 
                       row[1]['Isolate2'], 
                       time_delta=row[1]['Time Delta'], 
                       weight=row[1]['PWI'], 
                       trtype='Intra-Species')
        if row[1]['Host Species1'] != row[1]['Host Species2']:
            G.add_edge(row[1]['Isolate1'], 
                       row[1]['Isolate2'], 
                       time_delta=row[1]['Time Delta'], 
                       weight=row[1]['PWI'], 
                       trtype='Inter-Species')
```

```python
# Prune away any edges not equal to max_weight.
# Add those discarded edges to a separate graph that has 

for node in G.nodes(data=True):
    G_discarded.add_node(node[0], attr_dict=node[1])
    edges = G.in_edges(node[0], data=True)
    if len(edges) > 0:
        max_weight = max([edge[2]['weight'] for edge in edges])
        for edge in edges:
            if edge[2]['weight'] < max_weight:
                G.remove_edge(edge[0], edge[1])
                G_discarded.add_edge(edge[0], edge[1], attr_dict=edge[2])
```

```python
# Custom function to convert serialized datetime back into date object for comparison

def to_datetime(datestring):
    from datetime import date
    """
    e.g. '2010-08-12T00:00:00' --> date(2010, 08, 12)
    """
    
    y = int(datestring.split('-')[0])
    m = int(datestring.split('-')[1])
    d = int(datestring.split('-')[2].split('T')[0])
    
    return date(y,m,d)

to_datetime('2010-08-12T00:00:00')
```

```python
# 12 September 2014:
# A modification that we would like to do is as follows:
# - Find the earliest isolate in each subgraph.
# - Find the source for each of these earliest isolates in each subgraph.

# Here, I write custom functions to get that done.


# Function for finding earliest isolate in each subgraph:
def find_earliest_nodes(subgraph):
    """
    Returns one isolate that has a date equal to the earliest date amongst all of the 
    isolates in the subgraph.
    
    NOTE: If there are multiple isolates that fulfill the criteria, all of them are considered.
    """
    
    nodes = subgraph.nodes(data=True)
    
    dates = [node[1]['isolation_date'] for node in nodes]
    
    min_date = min(dates)
    
    earliest_nodes = []
   
    for node in nodes:
        if node[1]['isolation_date'] == min_date:
            earliest_nodes.append(node)
            
    return earliest_nodes
            
# Function for adding back the edge for a given earliest_node.
def find_source_node(node, discarded_edges_graph):
    """
    Basic idea of logic:
    1. In discarded_edges_graph, identify in_edges if source node's isolation_date < sink node's isolation_date 
       by more than 6 days.
    2. Identify edge(s) of max weight.
    3. Add it to a list of edges to be added back into the graph.
    
    NOTES:
    - node has to have have metadata provided as well. Otherwise, the code doesn't work. 
    """
    in_edges = [edge for edge in discarded_edges_graph.in_edges(node[0], data=True) if edge[2]['time_delta'] > 6]
    edges_to_add = []
    if len(in_edges) > 0:
        for edge in in_edges:
            source = edge[0]
            sink = edge[1]
            data = edge[2]
            maxweight = max([edge[2]['weight'] for edge in in_edges]) 
            if edge[2]['weight'] == maxweight:
                edges_to_add.append((source, sink, data))
                
    return edges_to_add
```

```python
# Find all subgraphs of G, and then add in edges that explain the earliest isolate of each subgraph.
# Prior to adding them back in, there should be 539 subgraphs.

subgraphs = list(nx.connected_component_subgraphs(G.to_undirected()))
print(len(subgraphs))

# for subgraph in subgraphs:
#     if len(subgraph.nodes()) == 1:
#         print(subgraph.nodes())

edges_to_add = []

for i in range(len(subgraphs)):
    earliest_nodes = find_earliest_nodes(subgraphs[i])
    
    for earliest_node in earliest_nodes:
    
        new_edges = find_source_node(earliest_node, G_discarded)    
        if len(new_edges) > 0:
            edges_to_add.extend(new_edges)
print(edges_to_add)
G.add_edges_from(edges_to_add)
```

```python
# This is a custom function for serializing dates such that it is JSON compatible. 
# It gets used right at the last step prior to the JSON dump.
from dateutil.tz import tzutc

UTC = tzutc()

def serialize_date(dt):
    """
    Serialize a date/time value into an ISO8601 text representation
    adjusted (if needed) to UTC timezone.

    For instance:
    >>> serialize_date(datetime(2012, 4, 10, 22, 38, 20, 604391))
    '2012-04-10T22:38:20.604391Z'
    """
    #if dt.tzinfo:
    #    dt = dt.astimezone(UTC)
    return dt.isoformat()
```

```python
# Dump the graph as a JSON node-link file.
# Prior to doing this, serialize all of the node dates to be JSON serializable.
G_serialized = deepcopy(G)

for node in G_serialized.nodes(data=True):
    node[1]['isolation_date'] = serialize_date(node[1]['isolation_date'])

d = json_graph.node_link_data(G_serialized) # node-link format to serialize
json.dump(d, open('viz/force2.json','w'))
```

```python
# Create a graph based on geography

geonet = nx.DiGraph()

year = df_full

# Add in the nodes 

geographic_locations = np.unique(year[['State1', 'State2']].values)
print(geographic_locations)

for place in geographic_locations:
    geonet.add_node(place)

# Add in the edges with initial weights 0

for row in year.iterrows():
    geonet.add_edge(row[1]['State1'], row[1]['State2'], weight=0)

# Add in the weights

for row in year.iterrows():
    if geonet.has_edge(row[1]['State1'], row[1]['State2']):
        geonet[row[1]['State1']][row[1]['State2']]['weight'] += row[1]['Weight']
        
# Add in lat-long data for each state
for row in lat_long_vals.iterrows():
    geonet.node[row[1]['State']]['N'] = row[1]['N']
    geonet.node[row[1]['State']]['W'] = row[1]['W']

fig = plt.figure(figsize=(12,12))
nx.draw_networkx(geonet)

d = json_graph.node_link_data(geonet) # node-link format to serialize
json.dump(d, open('viz/geonet.json','w'))
```
