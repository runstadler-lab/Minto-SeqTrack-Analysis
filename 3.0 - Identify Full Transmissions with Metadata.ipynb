```python
full_transmission = pd.DataFrame(transmission_matrix.stack()).reset_index()
#transmission_list.sort()
#full_transmission = transmission_list[transmission_list[0] > (0.] #threshold for HA gene only=0.959292
full_transmission.columns = ['Strain 1', 'Strain 2', 'PWI']
full_transmission.sort('PWI', ascending=True).head(10)
```

```python
masterlist = pd.read_csv('Raw Data/20140211 Minto and North America Combined.csv', index_col=0, parse_dates = ['Collection Date'])
# print len(masterlist)
# print len(masterlist)
for row in masterlist.iterrows():
    strain_name = row[1]['Strain Name'].split('(')
    row[1]['Strain Name'] = strain_name[0]
# masterlist = masterlist[(masterlist['Host Species'] != 'IRD:Human') & (masterlist['Host Species'] != "IRD:Swine")]
# masterlist = masterlist[masterlist['Collection Date'] < '2012']

# masterlist = masterlist.set_index('Strain Name')
masterlist
```

```python
ft2 = full_transmission.set_index('Strain 1').join(masterlist[['Collection Date', 'Host Species', 'Subtype', 'Country', 'State/Province']], how='inner', rsuffix='1')
#ft2 = ft2.join(seg4_transmissions.set_index('Strain 1')['PWI'], how='outer')
ft2
```

```python
ft3 = ft2.reset_index().set_index('Strain 2').join(masterlist[['Collection Date', 'Host Species', 'Subtype', 'Country', 'State/Province']], how='inner', rsuffix='2')
#ft3
ft3 = ft3.reset_index()
```

```python
ft3
```

```python
len(ft3)
```

```python
ft4 = ft3[(ft3['level_0'] != ft3['index']) & (ft3['PWI'] > 7.00)]
# ft4 = ft3[(ft3['level_0'] != ft3['index']) & (ft3['PWI'] > 7.00) & (ft3['Host Species'] != 'IRD:Human') & (ft3['Host Species'] != 'IRD:Swine') & (ft3['Host Species2'] != 'IRD:Human') & (ft3['Host Species2'] != 'IRD:Swine')]
ft4.head()
```

```python
len(ft4)
```

```python
ft4.columns = ['Isolate2', 'Isolate1', 'PWI', 'Collection Date1', 'Host Species1', 'Subtype1', 'Country1', 'State/Province1', 'Collection Date2', 'Host Species2', 'Subtype2', 'Country2', 'State/Province2']
ft4
```

```python
ft4['Time Delta'] = ft4['Collection Date2'] - ft4['Collection Date1']
ft4
```

```python
ft5 = ft4[(ft4['Time Delta'] > -7) & ((ft4['State/Province2'] == 'Alaska') | (ft4['State/Province1'] == 'Alaska'))]
ft5 = ft5.replace(to_replace=['MALL', 'NOPI', 'RNDU', 'AGWT'],
            value=['IRD:Mallard/Avian', 'IRD:Northern Pintail/Avian', 'IRD:Ring-necked Duck/Avian', 'IRD:American Green-winged Teal/Avian'])
ft5
```

```python
for row in ft5.iterrows():
    r = row[0]
    if row[1]['State/Province1'] == '-N/A-':
        state = row[1]['Isolate1'].split('/')[2]
        print state
        ft5.loc[r, 'State/Province1'] = state
        
        
    if row[1]['State/Province2'] == '-N/A-':
        state = row[1]['Isolate2'].split('/')[2]
        print state
        ft5.loc[r, 'State/Province2'] = state
```
