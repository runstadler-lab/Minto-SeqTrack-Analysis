```python
from sklearn.cluster import AffinityPropagation
import pandas as pd
import numpy as np
```

```python
segment = 4
```

```python
distmat = pd.read_csv('1.1 - Distmats/Segment %d Distmat.txt' % segment, delimiter=' ', skiprows=[0], index_col=0, header=None)
distmat.columns = distmat.index
distmat
```

```python
affmat = 1 - distmat
affmat.to_csv('2.0 - Affmats/Segment %d Affmat.csv' % segment)
affmat
```

```python
ap = AffinityPropagation(affinity='precomputed')
ap.fit(affmat)
```

```python
ap.labels_
```

```python
clusters = zip(affmat.index, ap.labels_)
clusters = pd.DataFrame(clusters)
clusters = clusters.set_index(0)
clusters.columns = ['Cluster']
clusters
```

```python
minvals = []
medianvals = []
for group in clusters.groupby('Cluster'):
    accession_numbers = [str(item) for item in group[1].index]
    #print accession_numbers
    
    subset = affmat[accession_numbers]
    subset = subset.loc[accession_numbers,:]
    
    #print group[0]
    minvals.append(np.matrix(subset).min())
    medianvals.append(np.median(subset))
    
print min(minvals)
```

```python
with open('2.0 - Minimum PWI/Segment %d Minimum PWI.txt' % segment, 'w+') as f:
    f.write('%s' % min(minvals))
```

```python
with open('2.0 - Median PWI/Segment %d Median PWI.txt' % segment, 'w+') as f:
    f.write('%s' % min(medianvals))
```
