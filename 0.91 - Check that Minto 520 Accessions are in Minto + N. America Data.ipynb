```python
import pandas as pd
from Bio import SeqIO
```

```python
# Read in the accession numbers for the Minto Flats isolates.
accessions = pd.read_csv('Raw Data/Minto Flats 520 Accession Numbers.csv', header=None)
accessions.values
```

```python
len(accessions)
```

```python
# Read in the full data set from the IRD
all_info = pd.read_csv('Raw Data/20140922 Minto Plus North America (2008-2011).csv', index_col='Sequence Accession', parse_dates=['Collection Date'])
all_info['Strain Name'] = all_info['Strain Name'].str.split('(').str[0]
all_info
```

```python
len(all_info) / 8
```

```python
# This is the list of accession numbers from the Minto Flats dataset
accessions_list = []
for row in accessions.values:
    for item in row:
        accessions_list.append(item)
```

```python
print(len([accession for accession in accessions_list if accession != '0']))
```

```python
# Check to see how many accession numbers are not present in all_info
not_in = []
for accession in accessions_list:
    if accession not in all_info.index:
        not_in.append(accession)
#         print(accession)

print(len(not_in)/8)
```

```python
# Some of the minto flats isolates cannot be found in the IRD by doing a strain name search. We want to know what they are.
```

```python
# These are the Interior Alaska strains from the IRD-downloaded dataset.
# print(len(all_info[all_info['Strain Name'].str.contains('Interior Alaska')])/8)
interior_alaska_strains = all_info[all_info['Strain Name'].str.contains('Interior Alaska')]
interior_alaska_strains
```

```python
# And these are the names of those strains.
ird_minto_strains = set(interior_alaska_strains['Strain Name'].values)
ird_minto_strains
```

```python
# These are the names of the Minto Flats isolates as obtained from Nichola
minto_names = pd.read_csv('Raw Data/Minto Flats 520 Names Cleaned.csv', index_col=0)
minto_names
```

```python
# This set of strains comes from Nichola's dataset
nic_minto_set = set(minto_names['Strain Name'].values)
```

```python
nic_minto_set.difference(ird_minto_strains)
```

```python
# These are the strains that are found in Nichola's set but cannot be found in the IRD dataset.
leftover_names = pd.DataFrame(list(nic_minto_set.difference(ird_minto_strains)))
leftover_names.columns = ['Strain Name']
leftover_names
```

```python
leftover_names.to_csv('Raw Data/Minto Flats Leftover Names.csv')
```
