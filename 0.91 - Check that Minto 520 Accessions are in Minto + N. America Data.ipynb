```python
# Some of the minto flats isolates cannot be found in the IRD by doing a strain name search. We want to know what they are.
```

```python
minto = pd.read_csv('20140206 Minto Flats Isolates.csv')
minto.dtypes
```

```python
minto_strains = []
for strain in minto.groupby('Strain Name'):
    #print strain[0].split('(')[0]
    minto_strains.append(strain[0].split('(')[0])
```

```python
minto_ird_set = set(minto_strains)
minto_ird_set
#print len(minto_ird_set)
```

```python
minto_names = pd.read_csv('Minto Flats 520 Names Cleaned.csv', header=None, delimiter=', ')
minto_names.T[0]
minto_nic_set = set(list(minto_names.T[0].values))
minto_nic_set
#print len(minto_nic_set)
```

```python
len(minto_nic_set.difference(minto_ird_set))
```

```python
with open('Minto Flats Leftover Names.csv', 'w+') as f:
    for item in minto_nic_set.difference(minto_ird_set):
        if item[-1] == ',':
            f.write('\n%s,' % item.split(',')[0])
        else:
            f.write('\n%s,' % item)
```
