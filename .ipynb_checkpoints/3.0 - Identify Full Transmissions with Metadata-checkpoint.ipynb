```python
import networkx as nx
import pandas as pd
```

```python
transmission_matrix = pd.read_csv('2.2 - Full Transmission/Estimated 8 Segment Transmissions.csv', index_col='Strain Name')
transmission_matrix
```

```python
transmission_matrix.values
```

```python
full_transmission = pd.DataFrame(transmission_matrix.stack()).reset_index()
full_transmission.columns = ['Strain 1', 'Strain 2', 'PWI']
full_transmission = full_transmission[full_transmission['Strain 1'] != full_transmission['Strain 2']]
# full_transmission = full_transmission[full_transmission['PWI'] > 7]
# full_transmission.sort('PWI', ascending=False)
full_transmission
```

```python
masterlist = pd.read_csv('Raw Data/20140211 Minto and North America Combined Dates Cleaned.csv', index_col=0, parse_dates = ['Collection Date'])
# print len(masterlist)
print(len(masterlist))
# for row in masterlist.iterrows():
#     strain_name = row[1]['Strain Name'].split('(')
#     row[1]['Strain Name'] = strain_name[0]
# masterlist = masterlist[(masterlist['Host Species'] != 'IRD:Human') & (masterlist['Host Species'] != "IRD:Swine")]
# masterlist = masterlist[masterlist['Collection Date'] < '2012']

masterlist = masterlist.set_index('Strain Name')
masterlist
```

```python
ft2 = full_transmission.set_index('Strain 1').join(masterlist[['Collection Date', 'Host Species', 'Subtype', 'Country', 'State/Province']], how='inner', rsuffix='1')
#ft2 = ft2.join(seg4_transmissions.set_index('Strain 1')['PWI'], how='outer')
ft2.index.rename('Isolate1', inplace=True)
ft2
```

```python
ft3 = ft2.reset_index()
# ft3
# ft3.columns
ft3 = ft3.set_index('Strain 2').join(masterlist[['Collection Date', 'Host Species', 'Subtype', 'Country', 'State/Province']], how='inner', rsuffix='2')
ft3.index.rename("Isolate2", inplace=True)
ft3 = ft3.reset_index()
ft3
```

```python
# This removes any self-transmissions
ft4 = ft3[(ft3['Isolate1'] != ft3['Isolate2'])]
# ft4 = ft3[(ft3['level_0'] != ft3['index']) & (ft3['PWI'] > 7.00) & (ft3['Host Species'] != 'IRD:Human') & (ft3['Host Species'] != 'IRD:Swine') & (ft3['Host Species2'] != 'IRD:Human') & (ft3['Host Species2'] != 'IRD:Swine')]
ft4
```

```python
ft4['Time Delta'] = ft4['Collection Date2'] - ft4['Collection Date']
ft4
```

```python
ft5 = ft4.replace(to_replace=['MALL', 'NOPI', 'RNDU', 'AGWT'],
            value=['IRD:Mallard/Avian', 'IRD:Northern Pintail/Avian', 'IRD:Ring-necked Duck/Avian', 'IRD:American Green-winged Teal/Avian'])
ft5
```

```python
# Sanity check: make sure that the number of nodes is still 1186
print (len(ft5))
print (len(ft5.groupby('Isolate2')))
```

```python
# Properly impute state names from the strain names for the Source isolate.
# strain_names_splitted1 = 
# strain_names_splitted1.index = ft5.index
# strain_names_splitted1
ft5['State/Province1'] = ft5['Isolate1'].str.split('/').str[2]
ft5['State/Province1']
```

```python
# Properly impute state names from the strain names for the Sink isolate.
# strain_names_splitted2 = 
# strain_names_splitted2.index = ft5.index
ft5['State/Province2'] = ft5['Isolate2'].str.split('/').str[2]
ft5['State/Province2']
```

```python
# Replace a spelling error - Nunavet should be Nunavut
# Also, replace BC with British Columbia
# Finally, replace MB with Manitoba
# Finally, replace Delaware Bay with Delaware
ft5 = ft5.replace("Nunavet", "Nunavut")
ft5 = ft5.replace("BC", "British Columbia")
ft5 = ft5.replace("MB", "Manitoba")
ft5 = ft5.replace("Delaware Bay", "Delaware")
# ft5
```
