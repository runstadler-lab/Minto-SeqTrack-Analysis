```python
import pandas as pd
```

```python
names = pd.read_csv('Raw Data/Minto Flats 520 Strain Names.csv', header=None)
names[0].values
```

```python
cleaned_names = []
for name in names[0].values:
    strain_name = name.split('(')[0]
    subtype = name.split('(')[1].split(')')[0]
    cleaned_names.append((strain_name, subtype))
```

```python
cleaned_names
```

```python
names_df = pd.DataFrame(cleaned_names)
names_df.columns = ['Strain Name', 'Subtype']
names_df.to_csv('Raw Data/Minto Flats 520 Names Cleaned.csv')
```
