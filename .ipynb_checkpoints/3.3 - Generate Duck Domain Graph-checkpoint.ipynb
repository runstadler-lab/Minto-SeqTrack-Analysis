```python
for sc, sk, d in G.edges(data=True):
    sc_5season = G.node[sc]['5season']
    sk_5season = G.node[sk]['5season']
    
    if sc.split('/')[1] != G.node[sc]['species']:
        print(sc, G.node[sc]['species'])
    
    if d['5season_transmission'] != '{0}-{1}'.format(sc_5season, sk_5season):
        print(d['5season_transmission'], sc, sc_5season, sk, sk_5season)
        G.edge[sc][sk]['5season_transmission'] = '{0}-{1}'.format(sc_5season, sk_5season)
```

```python
def has_years(g, years_iterable):
    truth = False
    for n, d in g.nodes(data=True):
        if d['isolation_date'].year in years_iterable:
            truth = True
    return truth
```

```python
def get_node(G, name_subset):
    for node in G.nodes():
        if name_subset in node:
            return node
        
get_node(G, '10BM06523')
```

```python
# Add weighting by cloacal/OP swabs - cloacal and OP swabs from the same bird
# share the same source, they should be weighted at half, otherwise 1.
nodes_with_cloral = [n for n, d in G.nodes(data=True) if d['oral_id'] != None or d['cloacal_id'] != None]

for node1, data in G2.nodes(data=True):
    if data['oral_id'] != None or data['oral_id'] != None:
        node2 = get_node(G2, data['oral_id'])
        if node2 == node1:
            node2 = get_node(G2, data['cloacal_id'])
        if node2 != None:
            # Get overlapping set of predecessors.
            node1_predecessors = set(G2.predecessors(node1))
            node2_predecessors = set(G2.predecessors(node2))
            intersect = node1_predecessors.intersection(node2_predecessors)

            if len(intersect) > 0:
                for source in intersect:
                    G2.edge[source][node1]['weighting_cloacal_op'] = 0.5
                    
for sc, sk, d in G2.edges(data=True):
    if 'weighting_cloacal_op' not in d.keys():
        G2.edge[sc][sk]['weighting_cloacal_op'] = 1
```

```python
def is_same_bird(G, node1, node2):
    band1 = G.node[node1]['band_num']
    band2 = G.node[node2]['band_num']
    
    web1 = G.node[node1]['web_tag']
    web2 = G.node[node2]['web_tag']
    
    # The only conditions under which True is returned is when:
    # 1. band1 and band2 are equal and both are not None, OR
    # 2. web1 and web2 are equal and both are not None.
    # Other conditions should return false.
    
    if band1 == band2 and band1 != None and band2 != None:
        return True
    elif web1 == web2 and web1 != None and web2 != None:
        return True
    else:
        return False
```

```python
# Add weighting by duck - are there edges that connect the same duck? They should be zero.
counter = 0 
time_deltas = []
for sc, sk, d in G2.edges(data=True):
    if is_same_bird(G2, sc, sk):
        G2.edge[sc][sk]['weighting_duck'] = 0
        counter += 1
        print(sc, sk)
        print(G2.node[sc]['band_num'], G2.node[sk]['band_num'])
        print(G2.node[sc]['web_tag'], G2.node[sk]['web_tag'])
        print(d['time_delta'])
        time_deltas.append(d['time_delta'].days)
    else:
        G2.edge[sc][sk]['weighting_duck'] = 1
        
print(counter)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.bar(Counter(time_deltas).keys(), Counter(time_deltas).values())
```

```python
# Finally, grab the minimum of all weightings - by timedelta, duck, and cloacal_op.
for sc, sk, d in G2.edges(data=True):
    weighting_duck = d['weighting_duck']
    weighting_cloacal_op = d['weighting_cloacal_op']
    weighting_timedelta = d['weighting_timedelta']
    
    weighting_minimum = min([weighting_duck, weighting_cloacal_op, weighting_timedelta])
    
    G2.edge[sc][sk]['weighting_minimum'] = weighting_minimum
```
