```python
# Load the Minto Flats pickle
G = nx.read_gpickle('Minto Flats.pkl')
G.edges(data=True)
```

```python
G.nodes(data=True)
```

```python
# How many nodes come from Interior Alaska?
alaskan_nodes = [n for n, d in G.nodes(data=True) if d['state'] == 'Interior Alaska']
len(alaskan_nodes)
```

```python
# How many interior alaska nodes were from 2009 and 2010
nodes = [n for n, d in G.nodes(data=True) if d['state'] == 'Interior Alaska' and d['isolation_date'] < datetime(2010, 9, 30) and d['isolation_date'] > datetime(2009, 3, 31)]
len(nodes)
```
