### Goal:
Get strain name in one column, and then 8 more columns of accession numbers for each segment.

```python
import pandas as pd
```

```python
america = pd.read_csv('Raw Data/20140922 Minto Plus North America (2008-2011).csv', na_filter=False, parse_dates=['Collection Date'])
america = america[(america['Host Species'] != "IRD:Human") & (america['Host Species'] != "IRD:Swine")]
america
```

```python
america_name = america['Strain Name']
names = pd.DataFrame(america_name.str.split('(').tolist())
names.columns = ['Strain Name', 'Subtype']
names.index = america.index
names
america['Strain Name'] = names['Strain Name']
america
```

```python
america_df = pd.DataFrame()
for group in america.groupby('Segment'):
    print group[0], len(group[1])
    america_df = america_df.join(group[1].set_index('Strain Name')['Sequence Accession'], rsuffix=group[0], how='outer')
```

```python
america_data = america[['Strain Name', 'State/Province', 'Country', 'Host Species', 'Collection Date', 'Subtype', 'Complete Genome']]
print len(america_data.drop_duplicates())
america_data = america_data.drop_duplicates()
america_data = america_data[(america_data['Host Species'] != "IRD:Human") & (america_data['Host Species'] != "IRD:Swine")]
print len(america_data)
america_data
```

```python
america_data.sort('Strain Name').drop_duplicates()
```

```python
america_df
```

```python
america_all = america_df.reset_index().drop_duplicates('Strain Name')
america_all = america_all.dropna().set_index('Strain Name').join(america_data.drop_duplicates('Strain Name').set_index('Strain Name'), how='outer')
print len(america_all)
```

```python
print len(america_all.dropna())
```
