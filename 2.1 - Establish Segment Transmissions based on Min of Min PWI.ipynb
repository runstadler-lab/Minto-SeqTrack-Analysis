```python
import pandas as pd
import numpy as np
```

```python
segment = 3
```

```python
segnums = [0, '', '2', '3', '4', '5', '6', '7', '8']
data = pd.read_csv('Raw Data/20140211 Minto and North America Combined.csv', index_col=0)
data = data[(data['Host Species'] != "IRD:Human") & (data['Host Species'] != "IRD:Swine")]
len(data)
# data[['Sequence Accession%s' % segnums[segment], 'Strain Name']]

data

for row in data.iterrows():
    strain_name = row[1]['Strain Name'].split('(')
    
    if row[1]['Subtype'] and len(strain_name) == 2:
        row[1]['Subtype'] = strain_name[1].split(')')[0]
    
#     if len(strain_name) > 1:
#         print strain_name
    row[1]['Strain Name'] = strain_name[0]
data.to_csv('Raw Data/20140211 Minto and North America Combined Name Cleaned.csv')


data
```

```python
affmat = pd.read_csv('2.0 - Affmats/Segment %d Affmat.csv' % segment, skiprows=[0], header=None, index_col=0)
affmat.columns = affmat.index
affmat
```

```python
with open('2.0 - Minimum PWI/Segment %d Minimum PWI.txt' % segment) as f:
    threshold = float(f.readlines()[0])
    print threshold
```

```python
info = data[['Sequence Accession%s' % segnums[segment], 'Strain Name']]
info = info.set_index('Sequence Accession%s' % segnums[segment])
info
```

```python
info = info.join(affmat).set_index('Strain Name')
info.columns = info.index
info
```
