```python
import numpy as np
import pandas as pd
```

```python
# This is the set of Interior Alaska isolates downloaded on 6 Feb 2014.
ird_minto = pd.read_csv('Raw Data/20140206 Minto Flats IRD Isolates Full Set.csv', parse_dates=['Collection Date'])
ird_minto
```

```python
nic_minto = pd.read_csv('Raw Data/Minto Flats 520 Sequences.csv', index_col='Sample ID', parse_dates=['Date'])
nic_minto
```

```python
for group in ird_minto.groupby('Segment'):
    nic_minto = nic_minto.join(group[1].set_index('Sample ID')['Sequence Accession'], rsuffix=group[0])

nic_minto = nic_minto.reset_index()
nic_minto
```

```python
len(nic_minto.drop_duplicates())
```

```python
cleaned_names = nic_minto['Common Name'].str.replace(')', '').str.split('(').tolist()
cleaned_names = pd.DataFrame(cleaned_names)
cleaned_names.columns = ['Common Name', 'Subtype']
# cleaned_names = cleaned_names.set_index()
cleaned_names
```

```python
nic_minto['Subtype'] = nic_minto['Common Name'].str.split("(").str[1].str.split(")").str[0]
nic_minto['Common Name'] = nic_minto['Common Name'].str.split('(').str[0]
nic_minto
```

```python
nic_minto.drop_duplicates().reset_index().set_index('Common Name').to_csv('Raw Data/20140206 Minto Flats 520 with Accessions.csv')
```
