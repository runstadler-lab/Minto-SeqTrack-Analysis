```python
import networkx as nx
import numpy as np
import pandas as pd
from networkx.readwrite import json_graph
import json
import datetime
from joblib import Parallel, delayed
```

```python
data = pd.read_csv('3.0 - Transmissions/Minto Plus North America Full Transmissions.csv', index_col=0, parse_dates=['Collection Date1', 'Collection Date2'])
data['Collection Date1'] = pd.to_datetime(data['Collection Date1'])
data['Collection Date2'] = pd.to_datetime(data['Collection Date2'])
data.dtypes
```

```python
G = nx.DiGraph()
```

```python
from dateutil.tz import tzutc

UTC = tzutc()

def serialize_date(dt):
    """
    Serialize a date/time value into an ISO8601 text representation
    adjusted (if needed) to UTC timezone.

    For instance:
    >>> serialize_date(datetime(2012, 4, 10, 22, 38, 20, 604391))
    '2012-04-10T22:38:20.604391Z'
    """
    #if dt.tzinfo:
    #    dt = dt.astimezone(UTC)
    return dt.isoformat()
```

```python
for row in data.iterrows():
    dsplit1 = str(row[1]['Collection Date1']).split(' ')[0].split('-')
    dsplit2 = str(row[1]['Collection Date2']).split(' ')[0].split('-')
    
    G.add_node(row[1]['Isolate1'], 
               host=row[1]['Host Species1'], 
               isolation_date=serialize_date(datetime.date(int(dsplit1[0]), int(dsplit1[1]), int(dsplit1[2]))),
               subtype=row[1]['Subtype1'],
               country=row[1]['Country1'],
               state=row[1]['State/Province1'])
    G.add_node(row[1]['Isolate2'], 
               host=row[1]['Host Species2'], 
               isolation_date=serialize_date(datetime.date(int(dsplit2[0]), int(dsplit2[1]), int(dsplit2[2]))),
               subtype=row[1]['Subtype2'],
               country=row[1]['Country2'],
               state=row[1]['State/Province2'])
    
    if row[1]['Host Species1'] == row[1]['Host Species2']:
        G.add_edge(row[1]['Isolate1'], row[1]['Isolate2'], time_delta=row[1]['Time Delta'], weight=row[1]['PWI'], trtype='Intra-Species')
    if row[1]['Host Species1'] != row[1]['Host Species2']:
        G.add_edge(row[1]['Isolate1'], row[1]['Isolate2'], time_delta=row[1]['Time Delta'], weight=row[1]['PWI'], trtype='Inter-Species')
```
